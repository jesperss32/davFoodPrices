```python
import pandas as pd
import numpy as np
import csv

production_df = pd.read_csv('Crop_Production_Normalized.csv', encoding='latin-1')
production_df
```

```python
# filter the df to include only production and exclude yield and area
production_df2 = production_df[production_df.Element == 'Production']
production_df2.head()
```

```python
# create list of countries in the price df
price_df = pd.read_csv('WFPVAM_FoodPrices_05-12-2017.csv', encoding='latin-1')

price_df.rename(columns={'adm0_id': 'country_ID', 'adm0_name': 'country', 'adm1_id' : 'district_ID', \
                   'adm1_name' : 'district', 'mkt_id' : 'market_ID', 'mkt_name' : 'market' , \
                   'cm_id' : 'product_ID','cm_name' : '_product', 'cur_id' : 'currency_ID', \
                   'cur_name' : 'currency', 'pt_id' : 'sale_ID', 'pt_name' : 'sale', 'um_id' : 'unit_ID', \
                   'um_name' : 'unit', 'mp_month' : 'month', 'mp_year' : 'year', 'mp_price' : 'price', \
                   'mp_commoditysource' : 'source'}, inplace=True)

price_df = price_df['country']

price_df

countries = list(set(price_df.tolist()))
print(countries)
```

```python
# filter production df to only include countries that are in the price df
production_df3 = production_df2[production_df2['Area'].isin(countries)]
production_df3
```

```python
# create list of products in price df
price_df = pd.read_csv('WFPVAM_FoodPrices_05-12-2017.csv', encoding='latin-1')

price_df.rename(columns={'adm0_id': 'country_ID', 'adm0_name': 'country', 'adm1_id' : 'district_ID', \
                   'adm1_name' : 'district', 'mkt_id' : 'market_ID', 'mkt_name' : 'market' , \
                   'cm_id' : 'product_ID','cm_name' : '_product', 'cur_id' : 'currency_ID', \
                   'cur_name' : 'currency', 'pt_id' : 'sale_ID', 'pt_name' : 'sale', 'um_id' : 'unit_ID', \
                   'um_name' : 'unit', 'mp_month' : 'month', 'mp_year' : 'year', 'mp_price' : 'price', \
                   'mp_commoditysource' : 'source'}, inplace=True)

price_df = price_df['_product']

price_products = list(set(price_df.tolist()))
print(price_products)
```

```python
# filter production df to only include products that are in the price df
production_df4 = production_df3[production_df3['Item'].isin(price_products)]
production_df4
```

```python
# create list of exact common products
production_df4_item = production_df4['Item']

common_products = set(production_df4_item.tolist())

print(sorted(list(common_products)))
print("Number of common products:")
print(len(common_products))
```

```python
# print lists of products from both dataframes,
# to inspect cases where diffent names are used for the same products
production_df3_item = production_df3['Item']

production_products = set(production_df3_item.tolist())

print("Number of products in production df:")
print(len(list(production_products)))

print("Products in the production df:")
for x in sorted(list(production_products)):
    print(x)
```

```python
print("Number of products in price df:")
print(len(price_products))

print("Products in the price df:")

for x in sorted(price_products):
    print(x)
```
